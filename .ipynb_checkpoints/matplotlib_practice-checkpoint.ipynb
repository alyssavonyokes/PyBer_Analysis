```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot. USING MATLAB BASED FUNCTIONS.
# Create plot and add error bar.
plt.bar(x_axis, y_axis, color="green", label='Boston', yerr=stdev, capsize=3)
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
# Print
plt.plot()
```

```python
# Create the plot (horozontal bar). USING MATLAB BASED FUNCTIONS.
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
# 'Get Current Axis' method to invert data and put Jan on top.
plt.gca().invert_yaxis()
# Add x and y axis labels.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
# Print
plt.plot()
```

```python
# Create the plot. USING OBJECT-ORIENTED INTERFACE METHOD.
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
# Add Legend.
ax.legend()
# Add Title.
ax.set_title("PyBer Fare by Month")
# Add lables to the x and y axis.
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
# Invert y axis labels.
ax.set_ylim("Dec", "Jan")
# Print.
plt.plot()
```

```python
# Create plot with error bar for OBJECT-ORIENTED APPROACH.
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Create scatter plot using .plot method in MATLAB.
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create scatter plot using .scatter method in MATLAB.
plt.scatter(y_axis, x_axis, c="red", label="Chicago")
# Add legend.
plt.legend()
# Add title and x and y axis labels.
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Limit the x axis data.
plt.xlim(0,50)
# Invert to place Jan at the top.
plt.gca().invert_xaxis()
# Print
plt.plot()
```

```python
# Create bubble chart. USING MATLAB BASED METHOD.
#Could've used this instead of for loop below --> plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
# Make y points larger.
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
# Make plot after the for loop.
plt.scatter(x_axis, y_axis, s=y_axis_larger)
# Print
plt.plot()
```

```python
# Scatter plot using the OBJECT-ORIENTED METHOD.
# Create plot and increase points by y axis size and times by 5 times the size (markers).
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, 
           s=[i * 5 for i in y_axis], 
           c="skyblue", alpha=0.2, 
           edgecolors="black", linewidth=2,
          label="Boston")
# Add legend.
ax.legend()
# Add title and x and y axis labels.
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
# Add a limit for the x axis data.
ax.set_ylim("Dec", "Jan")
# Print.
plt.plot()
```

```python
# Create pie plot using MATLAB INTERFACE.
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# Assign explode values.
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# Increase chart size.
plt.subplots(figsize=(8, 8))
# Create plot.
plt.pie(y_axis, 
        explode=explode_values, 
        labels=x_axis, 
        autopct='%.1f%%',
        colors=colors,)
# Print.
plt.show()
```

```python
# Create pie chart using OBJECT-ORIENTED INTERFACE.
# Add new colors. 
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# Explode values.
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# Create chart and format. Add start angle to set label Jan to top and reverse month order to clockwise direction. Add shadow.
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,labels=x_axis, 
       colors=colors, 
       explode=explode_values,
      autopct='%1.f%%',
      shadow="black",
      startangle=90,
      counterclock=False)
# Print.
plt.show()
```

```python
# Create chart and adjust major ticks to increment at every $5. MATLAB.
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# Create chart and adjust major ticks to increment at every $5. OBJECT-ORIENTED.
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
# Add Minor Ticks in-between.
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
